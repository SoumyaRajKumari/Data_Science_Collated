# Data Transforms

## Normalization

```python
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("winequality-red.csv")

## Method - Min Max Scaler
scaler = MinMaxScaler()
scaled = scaler.fit_transform(df[['fixed acidity']])

df['fixed acidity scaled'] = pd.Series(scaled.ravel()) 

figure, axis = plt.subplots(2)

axis[0].plot(df[['fixed acidity']])
axis[0].set_title("Before")

axis[1].plot(df[['fixed acidity scaled']])
axis[1].set_title("After")

plt.show()
```

## Standardization

```python
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("winequality-red.csv")

## Method - Standard Scaler
scaler = StandardScaler()
scaled = scaler.fit_transform(df[['fixed acidity']])

df['fixed acidity scaled'] = pd.Series(scaled.ravel()) 

figure, axis = plt.subplots(2)

axis[0].plot(df[['fixed acidity']])
axis[0].set_title("Before")

axis[1].plot(df[['fixed acidity scaled']])
axis[1].set_title("After")

plt.show()
```

## Robust Scaling

```python
from sklearn.preprocessing import RobustScaler
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("winequality-red.csv")

## Checking for Outliers
fig = plt.figure(figsize=(20, 5))
ax = fig.add_subplot(111)
ax.set_xticklabels(df.columns)
plt.boxplot(df)
plt.show()

## Method - Robust Scaler
scaler = RobustScaler()
scaled = scaler.fit_transform(df[['total sulfur dioxide']])

df['total sulfur dioxide scaled'] = pd.Series(scaled.ravel()) 

figure, axis = plt.subplots(2)

axis[0].plot(df[['total sulfur dioxide']])
axis[0].set_title("Before")

axis[1].plot(df[['total sulfur dioxide scaled']])
axis[1].set_title("After")

plt.show()
```

# Box-Cox Transformation

```python
from sklearn.preprocessing import PowerTransformer
import matplotlib.pyplot as plt

df = pd.read_csv("winequality-red.csv")

pt = PowerTransformer(method='box-cox')

df['total sulfur dioxide transformed'] = pd.Series(pt.fit_transform(df[['total sulfur dioxide']]).ravel())

figure, axis = plt.subplots(2)

axis[0].plot(df[['total sulfur dioxide']])
axis[0].set_title("Before")

axis[1].plot(df[['total sulfur dioxide transformed']])
axis[1].set_title("After")

plt.show()
```

# Yeo-Johnson Transformation

```python
from sklearn.preprocessing import PowerTransformer
import matplotlib.pyplot as plt

df = pd.read_csv("winequality-red.csv")

pt = PowerTransformer(method='yeo-johnson')

df['total sulfur dioxide transformed'] = pd.Series(pt.fit_transform(df[['total sulfur dioxide']]).ravel())

figure, axis = plt.subplots(2)

axis[0].plot(df[['total sulfur dioxide']])
axis[0].set_title("Before")

axis[1].plot(df[['total sulfur dioxide transformed']])
axis[1].set_title("After")

plt.show()
```

# One Hot / Dummy Variable Encoding

```python
import pandas as pd

df = pd.read_csv("Titanic_Dataset.csv")

## drop_first=True to avoid dummy variable trap
encoded_data = pd.get_dummies(df, columns = ['Pclass'],drop_first=True)
print(encoded_data.head())
```

```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

df = pd.read_csv("Titanic_Dataset.csv")

encoder = OneHotEncoder(sparse=False)

encoded_data = pd.DataFrame(encoder.fit_transform(df[['Pclass']]))
print(encoded_data.head())
```

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

df = pd.read_csv("Titanic_Dataset.csv")

encoder = LabelEncoder()

encoded_data = pd.DataFrame(encoder.fit_transform(df[['Pclass']]))
print(encoded_data.head())
```
